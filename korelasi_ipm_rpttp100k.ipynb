#Cleaning Data Mentah

Untuk ipm buang data 'IPM 2022', dan untuk rpttp100k buang data 'RPTTP100K 2010'

```python
data_ipm_cor = data_ipm.drop(["IPM 2022"],axis = 1)
data_ipm_cor
```

```python
data_rpttp100k_cor = data_rpttp100k.drop(["RPTTP100K 2010"],axis = 1)
data_rpttp100k_cor
```

Ditemukan ada data kosong di data set RPTTP100K pada provinsi

1.   KALIMANTAN UTARA tahun 2017 - 2010

2.   SULAWESI BARAT tahun 2016 - 2010

3.   PAPUA BARAT tahun 2014 - 2010

Jadi data tahun tersebut akan di drop


#Gabung data

```python
df_korelasi_2021 = pd.concat([data_ipm_cor['IPM 2021'], data_rpttp100k_cor['RPTTP100K 2021']],axis=1)
df_korelasi_2020 = pd.concat([data_ipm_cor['IPM 2020'], data_rpttp100k_cor['RPTTP100K 2020']],axis=1)
df_korelasi_2019 = pd.concat([data_ipm_cor['IPM 2019'], data_rpttp100k_cor['RPTTP100K 2019']],axis=1)
df_korelasi_2018 = pd.concat([data_ipm_cor['IPM 2018'], data_rpttp100k_cor['RPTTP100K 2018']],axis=1)
```
